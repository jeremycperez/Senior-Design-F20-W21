```python
## 0 0.253125 0.663542 0.350000 0.443750
## 0 0.380469 0.226042 0.242188 0.364583
import cv2
import os

path = r'C:\Users\jpere\Pictures\yoloSamples'
os.chdir(path)

dataset_text = 'datasetYOLO.txt'

pix = [640,480]

def YOLO_to_COCO(size, coord):
    x2 = int(((2*size[0]*float(coord[0]))+(size[0]*float(coord[2])))/2)
    x1 = int(((2*size[0]*float(coord[0]))-(size[0]*float(coord[2])))/2)

    y2 = int(((2*size[1]*float(coord[1]))+(size[1]*float(coord[3])))/2)
    y1 = int(((2*size[1]*float(coord[1]))-(size[1]*float(coord[3])))/2)
    
    w = x2-x1
    h = y2-y1
    print("(x1, y1, x2, y2)  :  ",x1,y1,x2,y2)
    print("    (x, y, w, h)  :  ",x1,y1,w,h)
    return (x1,y1,w,h)

count = 1
start = count

while(count <= 600):
    file = 'img%d.txt' % count
    text = open(file, 'r') 
    out = text.readlines();
    #print(out)
    text.close()
    print("\t\t",file)
    #print("\n\tLEN OUT:   \n", len(out))

    textList = []
    for i in range(len(out)):
        out[i] = str(out[i])
        textList += [out[i]]
    #c = out.split(' ')
    #classID.append(c)
    #print("\n\tTEXT LIST:  \n",textList)

    data = []

    classID = []
    num = ""
    for i in range(len(textList)):
        temp = []
        #print("\n\ti RANGE OF TEXT LIST:\n", len(textList[i]))
        for j in range(len(textList[i])):
            #print(j)
            #print("num: ",num)
            if(j == 0):
                classID.append(textList[i][j])
            elif(j == 1):
                pass
            elif(textList[i][j] != '\n'):
                if(textList[i][j] == ' '):
                    temp+=[float(num)]
                    num = ""
                else:
                    num+=(str(textList[i][j]))
            elif(textList[i][j] == '\n'):
                temp+=[float(num)]
                num = ""
                data.append(temp)



    print("\n\tCLASS ID:  \n",classID)
    #yolo = [ybb1, ybb2]
    print("\n\tBBOX EXTRACTION:\n",data)
    print("\n\tBBOX YOLO TO COCO")
    
    dataStr = []
    for i in range(len(data)):
        data[i] = YOLO_to_COCO(pix, data[i])
        dataStr += classID[i] + ','
        dataStr+=str(data[i])+'$'
        

    
    print("\n\tDataStr\n", dataStr)
    
    bbox = ''.join(dataStr)
    bbox = bbox.replace(')', '')
    bbox = bbox.replace('(', '')
    bbox = bbox.replace(' ', '')
    bbox = bbox.replace('$', ' ')
    print("\n\tBBOX STRING\n", bbox)
    
    file = r'img%d.jpg' % count
    #file = r'img599.jpg'
    
    imgName = 'img%d.jpg' % count
    imgPath = '/home/jeremy/yolov4-conv/data/img/'
    
    
    if(start == count):
        with open(dataset_text, 'w') as f:
            f.write(imgPath+imgName+' '+bbox+'\n')
    else:
        with open(dataset_text, 'a') as f:
            f.write(imgPath+imgName+' '+bbox+'\n')
    
    count+=1
    
    '''
    ref = cv2.imread(file)
    

    ###
    while True:
        for i in range(len(data)):
            if i == 0:
                bbox = cv2.rectangle(ref, (data[i][0], data[i][1]), (data[i][2]+data[i][0], data[i][3]+data[i][1]), (0, 0, 255), 1)
            else:
                bbox = cv2.rectangle(bbox, (data[i][0], data[i][1]), (data[i][2]+data[i][0], data[i][3]+data[i][1]), (0, 0, 255), 1)
        cv2.imshow("img",bbox)

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    count+=1
    ###
    
cv2.destroyAllWindows() 
'''
WINDOWS_LINE_ENDING = b'\r\n'
UNIX_LINE_ENDING = b'\n'

with open(dataset_text, 'rb') as open_file:
    content = open_file.read()

content = content.replace(WINDOWS_LINE_ENDING, UNIX_LINE_ENDING)

with open(dataset_text, 'wb') as open_file:
    open_file.write(content)
# cv2.imshow(bbox)
```
